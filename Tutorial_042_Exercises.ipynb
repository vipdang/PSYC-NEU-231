## Read in data from eeg study
* This time I've kept in data across a 1s window from many electrodes (20) sampled at 250Hz, so now we have a [960, 20, 250] matrix that contains data from all 960 trials, 20 electrodes, and over a 1s window following stimulus onset
* Also read in a "cond_labels" vector that has 960 0's and 1's to mark trials from the two experimental conditions 

## import stuff...

```python
import numpy as np
from scipy.linalg import eigh
from scipy.linalg import sqrtm

import matplotlib.pyplot as plt

fig_font = {'fontname':'Arial', 'size':'20'}

from sklearn.svm import SVC 
```

```python
# load the data...
eeg = np.load("class_demo_eeg.npz")

# get the different arrays like this...
eeg['data']
eeg['sr']
eeg['tx']
eeg['cond_labels']

# shape of data
print('Shape of the big eeg data set: ', eeg['data'].shape)

# and if you want to save some typing, especially because we only have a few variables, you reassign the different arrays like this
data = eeg['data']
sr = eeg['sr']
tx = eeg['tx']
cond_labels = eeg['cond_labels']
```

## Compute the mean across the first dimension (trials) and then plot the mean response across time in each of the 20 electrodes on one axis

* Might need to transpose the data into a Time x Electrode format

```python
mean_of_data = np.mean(data, axis=0)
print(mean_of_data.shape)
plt.plot (tx, mean_of_data.T)
plt.xlabel ('Time', **fig_font)
plt.ylabel ('Electrode', **fig_font)
plt.show ()
```

## That gives you an idea about what the pattern of data looks like across electrodes. 
* Next, take the mean across time, so you have a 960 x 20 matrix 
* Use the cond_labels vector to sort the data based on condition (so you have two 480 x 20 matrices)
* Do classification to see if you can decode the experimental condition
* Remember - break up your data into independent training and testing sets
* Set C to a small number if this is taking a long time...

```python
mean_time = np.mean(data, axis=2)
data_cond_0 = mean_time[cond_labels==0,:]
data_cond_1 = mean_time[cond_labels==1,:]

percent_training = 0.9
num_training_per_cond = int(np.floor(percent_training * data_cond_0.shape[0]))

training_data = np.vstack( (data_cond_0[:num_training_per_cond,:], data_cond_1[:num_training_per_cond,:]) )
training_labels = np.hstack((np.zeros(num_training_per_cond), np.ones(num_training_per_cond)))

test_data = np.vstack((data_cond_0[num_training_per_cond:,:],data_cond_1[num_training_per_cond:,:]))
num_test_per_cond = int(test_data.shape[0]/2)
test_labels = np.hstack((np.zeros(num_test_per_cond), np.ones(num_test_per_cond)))

model = SVC(kernel='linear',C=1)
model.fit(training_data, training_labels)

labels = model.predict(test_data)
accuracy = np.sum(labels==test_labels) / test_data.shape[0]
print(accuracy)
```

## Now try doing this across timepoints!
* First do a time window based on where you think there is nice separation in the responses across electrodes (look at the data that you plotted above)...maybe try 300-400msec post stim
* Then you can try timepoint x timepoint or a moving window

```python
window = (tx>=300) & (tx<400)
total_data = np.mean(data[:,:,window],2)
data_0 = total_data[cond_labels==0,:]
data_1 = total_data[cond_labels==1,:]

percent_training = 0.9
num_training_per_cond = int(np.floor(percent_training*data_0.shape[0]))

training_data = np.vstack((data_0[:num_training_per_cond,:], data_1[:num_training_per_cond,:]))
training_labels = np.hstack((np.zeros(num_training_per_cond), np.ones(num_training_per_cond)))

test_data = np.vstack((data_0[num_training_per_cond:,:], data_1[num_training_per_cond:,:]))
num_testing_per_cond = int(test_data.shape[0]/2)
test_labels = np.hstack((np.zeros(num_testing_per_cond), np.ones(num_testing_per_cond)))

model = SVC(kernel='linear',C=1)
model.fit(training_data,training_labels)
labels = model.predict(test_data)
accuracy = np.sum(labels==test_labels) / test_data.shape[0]
print(accuracy)
```
